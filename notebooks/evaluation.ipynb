# Evaluation Notebook

Notebook minimo para cargar el dataset de evaluacion, ejecutar el script de evaluacion RAGAS-style y mostrar las metricas principales.

```python
from pathlib import Path
import json

dataset_path = Path("notebooks/eval_dataset.json")
dataset = json.loads(dataset_path.read_text(encoding="utf-8"))
print(f"Casos cargados: {len(dataset)}")
dataset[:2]
```

```python
import subprocess
import sys
from pathlib import Path

command = [
    sys.executable,
    "scripts/ragas_eval.py",
    "--dataset",
    "notebooks/eval_dataset.json",
    "--output",
    "reports/ragas_report.json",
]
result = subprocess.run(command, capture_output=True, text=True)
print(result.stdout)
if result.stderr:
    print(result.stderr)

report_path = Path("reports/ragas_report.json")
if not report_path.exists():
    raise RuntimeError("No se genero reports/ragas_report.json")

print(f"Return code: {result.returncode}")
```

```python
report = json.loads(Path("reports/ragas_report.json").read_text(encoding="utf-8"))
metrics = report["metrics"]
metrics
```
